```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Deixa os gráficos aparecerem direto no notebook
%matplotlib inline

# Troca pelo nome EXATO do seu arquivo
arquivo = "Telemedicine_Use.csv"

df = pd.read_csv(arquivo)
df.head()
```

```python
df.columns

```

```python
# Renomear a coluna de percentual para um nome mais claro
df = df.rename(columns={"Value": "Pct_Telehealth"})

# Tirar linhas sem valor de telemedicina
df = df.dropna(subset=["Pct_Telehealth"])

# Garantir que Pct_Telehealth é número
df["Pct_Telehealth"] = pd.to_numeric(df["Pct_Telehealth"], errors="coerce")
df = df.dropna(subset=["Pct_Telehealth"])

# Calcular o percentil 75 (top 25%)
p75 = df["Pct_Telehealth"].quantile(0.75)
print("Percentil 75 de Pct_Telehealth:", p75)

# Criar classe binária: 1 = alta adoção, 0 = baixa adoção
df["alta_adocao"] = (df["Pct_Telehealth"] >= p75).astype(int)

# Ver primeiras linhas para conferir
df[["Pct_Telehealth", "alta_adocao"]].head()
```

```python
import matplotlib.pyplot as plt

# Histograma do percentual de telemedicina
df["Pct_Telehealth"].hist(bins=20)
plt.title("Distribuição de Pct_Telehealth")
plt.xlabel("Pct_Telehealth")
plt.ylabel("Frequência")
plt.show()

# Distribuição da classe alta_adocao
df["alta_adocao"].value_counts(normalize=True).plot(kind="bar")
plt.title("Proporção: baixa (0) x alta (1) adoção de telemedicina")
plt.xticks(rotation=0)
plt.show()
```

```python
# Quais colunas vamos usar como explicativas
cols_features_categoricas = ["Group", "State", "Subgroup", "Phase", "Time Period"]

# Garantir que todas existem no DF
cols_features_categoricas = [c for c in cols_features_categoricas if c in df.columns]
print("Features usadas:", cols_features_categoricas)

# One-hot encoding nas categóricas
X = pd.get_dummies(df[cols_features_categoricas], drop_first=True)

# Target
y = df["alta_adocao"]

X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Criar e treinar o modelo
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

# Predições
y_pred = log_reg.predict(X_test)
y_proba = log_reg.predict_proba(X_test)[:, 1]

print("Relatório de classificação:\n")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Matriz de Confusão – Alta vs Baixa Adoção de Telemedicina")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

fpr, tpr, thresholds = roc_curve(y_test, y_proba)
auc = roc_auc_score(y_test, y_proba)

plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Curva ROC – Modelo de Classificação")
plt.legend()
plt.tight_layout()
plt.show()
```
